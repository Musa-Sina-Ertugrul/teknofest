```python
# Define sentiment mapping
sentiment_mapping = {'Notr': 0, 'Negative': 1, 'Positive': 2}

# Function to pad sentences and convert to lowercase
def pad_sentence_lower(sentence, max_length=10, pad_token=0):
    words = re.sub(r'[^\w\s]', '', sentence).lower().split()[:max_length]  # Remove punctuation, convert to lowercase and split
    padded_words = words + [pad_token] * (max_length - len(words))
    return padded_words

pos_words = set(["harika", "mükemmel", "uygun", "çok iyi", "hızlı", "memnun", "yardımcı", "güler yüzlü", "kibar", "ilgili", "beklentilerimi", "harika",
                 "harika","inanılmaz","memnun","kesinlikle","kalite","güvenilir","kalitesi","ürün","hizmet","iyi","kargo","uygun","zamanında","şaka","hızlı",
                 "mükemmel","paketleme","fiyat","performans","net"])
neg_words = set(["kötü", "yetersiz", "hasarlı", "uzun", "ilgisiz", "pahalı", "yanıltıcı", "sorun", "zor", "kaba", "eksik", "berbat","düşük","yetersiz",
                 "hasarlı","zor","ilgisiz","açıklamalara","uymuyor","gecikti","resimlerden farklı","eksik","bozuk","kalitesiz","sorun","bozuk","kalitesiz",
                 "sorun","beklenen","uzun","sürpriz","olmadı","hayret","verici","değil"])
# Function to create reviews with only company names, in lowercase, and padded sentences with sentiment values
def create_reviews_with_padding_and_sentiments_simple_lower_all_companies(reviews, sentiments,companies):
    data = []
    for review,sentiment in zip(reviews,sentiments):
        for company in companies:
            full_review : str = review
            for punc in string.punctuation:
                full_review = full_review.replace(punc,"")
            splitted_review = full_review.casefold().split()
            for index in range(len(splitted_review)):
                splitted_review.insert(index,str(company))
                padded_review = pad_sentence_lower(full_review, max_length=10)
                sentiment_values = [sentiment_mapping['Notr']] * len(padded_review)  # Neutral padding
                for i, word in enumerate(re.sub(r'[^\w\s]', '', full_review).split()[:10]):
                    if any(pos_word in word for pos_word in pos_words):
                        sentiment_values[i] = sentiment_mapping['Positive']
                    elif any(neg_word in word for neg_word in neg_words):
                        sentiment_values[i] = sentiment_mapping['Negative']
                sentiment_values[index] = sentiment
                padded_review_with_sentiments = sentiment_values
                data.append({
                    'review': " ".join(splitted_review),
                    'sentiment': int(sentiment),
                    'padded_review': padded_review_with_sentiments
                })
        
    return data
```
