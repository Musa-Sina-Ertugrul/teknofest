```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
URL = "https://toptalent.co/yazilim-sirketleri"
r = requests.get(url=URL)
```

```python
soup = BeautifulSoup(r.content,"html5lib")
tbodies = soup.find_all("tbody")
print(tbodies[0].find_all("tr"))
```

```python
results = []
for tbody in tbodies:
    for trow in tbody.find_all("tr"):
        results.append(str(trow.find_all("td")[0].get_text()).casefold().strip())
```

```python
URL_2 = "https://www.netizenlist.com/fast-company-turkiyeye-gore-turkiyenin-en-buyuk-100-internet-sirketi-2023/"
r = requests.get(url=URL_2)
```

```python
soup = BeautifulSoup(r.content,"html5lib")
tbodies = soup.find_all("ul")
lis = tbodies[2].find_all("li")
```

```python
for li in lis:
    results.append(str(li.get_text()).casefold().split("–")[0].split("–")[0].strip())
results = list(set(results))
```

```python
df = pd.DataFrame(data=[*results],columns=["companies"])
```

```python
df.to_csv("./companies.csv",index=False)
```
